### Парсим инфобоксы из Википедии (wptools)

```python
infoboxes = []

for anthem in anthems.keys():
    name_en = anthems[anthem]['name_en']
    try:
        infobox = wptools.page(name_en).get_parse().data['infobox']
    except:
        try:
            infobox = wptools.page(name_en).get_parse().data['infobox']
        except:
            print(name_en, 'skipped')
    infoboxes.append(infobox)
```

```python
for infobox in infoboxes:
    
    if infobox != None:
        
        # state characteristics
        try:
            state_characteristics = re.findall('\[{2}\S+\s?\S*\]{2}', infobox['government_type'])
            state_characteristics = [characteristic.lstrip('[[').rstrip(']]').lower().split('|')[0] 
                                 for characteristic in state_characteristics]
            anthems[infoboxes.index(infobox)]['state_characteristics'] = state_characteristics
        except KeyError:
            print(infoboxes.index(infobox), anthems[infoboxes.index(infobox)]['name'], 'no government type')
        
        # religions
        try:
            religion = infobox['religion']
            
            if '%' in religion:
                religions_list = re.findall('\d+\.?\d*% \[*\D+\]*', religion)
                
                religions_dict = {}

                for religion in religions_list:
                    share = float(re.match('\d+\.?\d*', religion).group(0))
                    name = re.search(' \[*\'*\w+', religion).group(0).lower().lstrip(' [[')
                    religions_dict[name] = share

                anthems[infoboxes.index(infobox)]['religion'] = religions_dict
                
        except KeyError:
            print(infoboxes.index(infobox), anthems[infoboxes.index(infobox)]['name'], 'no religion')
        
        # area
        try:
            area = infobox['area_km2'].replace(',', '')
            area = re.search('\d+', area).group(0)
            area = int(area)
            anthems[infoboxes.index(infobox)]['area'] = area
        except KeyError:
            print(infoboxes.index(infobox), anthems[infoboxes.index(infobox)]['name'], 'no area')
        
        # population
        pop = 0
        try:
            population = infobox['population_estimate'].replace(',', '')
            pop =1
        except:
            try:
                population = infobox['population_census'].replace(',', '')
                pop = 1
            except:
                print(infoboxes.index(infobox), anthems[infoboxes.index(infobox)]['name'], 'no population')
            
        try:
            if pop == 1:
                population = re.search('\d+', population).group(0)
                population = int(population)
                anthems[infoboxes.index(infobox)]['population'] = population
        except AttributeError:
            print(infoboxes.index(infobox), anthems[infoboxes.index(infobox)]['name'], 'no population')
         
        # former colony or not
        if 'sovereignty_type' in infobox:
            anthems[infoboxes.index(infobox)]['former_colony'] = 1
        else:
            anthems[infoboxes.index(infobox)]['former_colony'] = 0
    else:
        print(infoboxes.index(infobox), anthems[infoboxes.index(infobox)]['name'], 'NO INFO')
```
